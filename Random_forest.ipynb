```julia
using CSV
using Statistics,DataFrames
using PlotlyJS
```

```julia
# read the CSV file
df = CSV.read("hikari.csv",DataFrame)
```

```julia

main_df = select(df, [:"idle.tot", :"flow_iat.tot", :"fwd_iat.tot", :"active.avg", :"idle.max", :"flow_iat.max", :"fwd_iat.max", :"active.max", :"idle.avg", :"active.tot", :"bwd_iat.max", :"idle.min", :"flow_iat.avg", :"fwd_iat.avg", :"flow_iat.min", :"fwd_iat.min", :"fwd_iat.std", :"bwd_iat.std", :"flow_iat.std", :"bwd_iat.avg", :"Label"])
```

```julia
using DecisionTree
using Random,Statistics
```

```julia
train_ratio = 0.8
test_ratio = 1.0 - train_ratio

nrows = nrow(main_df)
train_indices = randsubseq(1:nrows, train_ratio)
test_indices = setdiff(1:nrows, train_indices)

train_df = main_df[train_indices, :]
test_df = main_df[test_indices, :]
```

```julia
describe(train_df)
```

```julia
X_train = Matrix(select(train_df, Not(:Label)))

y_train = Vector(train_df.Label)

X_test = Matrix(select(test_df, Not(:Label)))

y_test = Vector(test_df.Label)
```

```julia
rf_model = RandomForestClassifier(max_depth=10, min_samples_leaf=10)
```

```julia
fit!(rf_model, X_train, y_train)
```

```julia
y_pred = predict(rf_model, X_test)
```

```julia
accuracy = mean(y_pred.==y_test)
```
